```python
def ngram_frequency(text, n, top):
  """
  text: input string
  n: i.e. bigrams, n = 2
  top: i.e. top 20
  """
  import nltk
  import string
  import pandas as pd

  # Tokenization
  tokens = nltk.word_tokenize(text)

  # Eliminate punctuation and stopwords
  stopwords = nltk.corpus.stopwords.words("english")
  tokens = [word.lower() for word in tokens if word.lower() not in stopwords and word not in string.punctuation and any(c.isalpha() for c in word)]
  ## any(c.isalpha() for c in word)
  ### "if any character in word is in the alphabet"
  #### This eliminates extraneous punctuation "===" but keeps words like "text-to-speech"

  ## Create ngrams
  ngrams_list = nltk.ngrams(tokens, n)

  ## Create frequency distribution
  ngrams_freq = nltk.FreqDist(ngrams_list).most_common(top)

  ngram_freq_dict = dict(ngrams_freq)

  ## Sort values by highest frequency
  ngram_sorted = {k:v for k,v in sorted(ngram_freq_dict.items(), key=lambda item:item[1])}

  # Join ngram tokens with '_' + maintain sorting
  ngram_joined = {'_'.join(k):v for k,v in sorted(ngram_freq_dict.items(), key=lambda item:item[1])}

  # Convert to Pandas series before plotting
  ngram_freq_pd = pd.Series(ngram_joined)

  return ngram_freq_pd
```
